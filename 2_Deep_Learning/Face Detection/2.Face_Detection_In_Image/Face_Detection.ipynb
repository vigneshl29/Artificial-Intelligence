```python
# Importing opencv library
import cv2
```

```python
# Loading the cascades
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# Defining the function for detections
def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
    return frame
```

### Image before detection

```python
# Importing matplotlib
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (5,5)
# Loading original image and plotting it
image_bgr = cv2.imread('vignesh.jpg',cv2.IMREAD_COLOR)
#convetging to rgb
original_image = cv2.cvtColor(image_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(original_image)
```

#### Detecting face in this image

```python
# Converting image to gray scale
gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
# Detecting face
detected_image = detect(gray, original_image)
```

### Image after detection

```python
# Plotting detected image
plt.imshow(detected_image)
```
